```python
import ee, geemap
```

```python
PROJECT_ID = "ee-abirnahraf"
```

```python
ee.Initialize(project='ee-abirnahraf')
```

```python
m = geemap.Map(center=[23.8103, 90.4125], zoom=9)
```

```python
srtm = ee.Image("USGS/SRTMGL1_003")
m.addLayer(srtm, {"min": 0, "max": 1000}, "SRTM Elevation")
```

```python
m.add_basemap("HYBRID")
m
```

```python
m.to_html("gee_map.html", title="Minimal GEE Map with Drawing Tools")
print("💾 Map saved as gee_map.html")
```
